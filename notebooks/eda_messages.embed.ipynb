## EDA on raw messages

### Message counts per participant

**Histogram on \# of messages per participant**

```
#| label: fig-message_count
#| fig-cap: "Distribution of Message Counts"
#| fig-width: 3.5
#| fig-height: 2.5
raw_data |> 
    group_by(subid) |> 
    summarize(n = n()) |> 
    ggplot(aes(x = n)) + 
    geom_histogram(bins = 50, fill = "#97C4C5") +
    labs(x = "# of messages", y = "subject counts")
```

### Message Length per participant

**Histogram on mean/median message length per participant**

```
#| label: fig-message_length
#| fig-cap: "Distribution of Mean Message Length"
#| fig-width: 3.5
#| fig-height: 2.5
raw_data |> 
  group_by(subid) |> 
  summarize(mean_length = mean(text_length)) |> 
  ggplot(aes(x = mean_length)) + 
  geom_histogram(binwidth = 1, fill = "#97C4C5") +
  labs(x = "mean message length", y = "subject counts")
```

```
raw_data |> 
  group_by(subid) |> 
  summarize(median_length = median(text_length)) |> 
  ggplot(aes(x = median_length)) + 
  geom_histogram(binwidth = 1, fill = "#97C4C5") +
  labs(x = "median message length", y = "subject counts")
```

## EDA on prediction windows

### prediction window: 3-day

**Distribution by label**

```
plt1 <- pred_3day |> 
  group_by(subid, day_start) |> 
  summarize(n = n()) |> 
  ggplot(aes(x = n)) + 
  geom_histogram(fill = "#97C4C5") +
  labs(x = "number of messages", y = "lapse counts", 
       caption = str_wrap("(a) Distribution of Message Counts"), width = 50) +
  theme(plot.caption = element_text(hjust = 0.5)) 
```

**Distribution by participant**

```
plt2 <- pred_3day |> 
  group_by(subid, day_start) |> 
  summarize(
    n_messages = sum(!na)
  ) |> 
  group_by(subid) |> 
  summarize(
    mean_messages = mean(n_messages)
  ) |>
  ggplot(aes(x = mean_messages)) +
  geom_histogram(fill = "#97C4C5") +
  labs(x = "averaged number of messages per label", y = "subject counts",
       caption = str_wrap("(b) Distribution of Message Counts by Subject",
                          width = 50)) +
  theme(plot.caption = element_text(hjust = 0.5)) 
```

```
plt3 <- pred_3day |> 
  group_by(subid, day_start) |> 
  summarize(
    missingness = sum(na) / n()
  ) |> 
  group_by(subid) |> 
  summarize(
    mean_missing = mean(missingness)
  ) |> 
  ggplot(aes(x = mean_missing)) +
  geom_histogram(fill = "#97C4C5") +
  labs(x = "percentage of missingness", y = "subject counts",
       caption = str_wrap("(c) Distribution of Missingness by Subject", 
                          width = 50)) +
  theme(plot.caption = element_text(hjust = 0.5)) 
```

```
#| label: fig-3day
#| fig-cap: "Sample Distribution in the 3-Day Prediction Window"
ggpubr::ggarrange(
  ggpubr::ggarrange(NULL, plt1, NULL, ncol = 3, widths = c(1, 1, 1)), # Centering plot1 in the first row
  ggpubr::ggarrange(plt2, plt3, ncol = 2, nrow = 1),               # Arranging plot2 and plot3 in the second row
  nrow = 2, ncol = 1, heights = c(1, 1) 
)
```

### prediction window: 1 week

**Distribution by label**

```
#| label: fig-message_count_1week
#| fig-cap: "Distribution of Message Counts in the 1-Week Prediction Window"
#| fig-width: 3.5
#| fig-height: 2.5
pred_1week |> 
  group_by(subid, day_start) |> 
  summarize(n = n()) |> 
  ggplot(aes(x = n)) + 
  geom_histogram(fill = "#97C4C5") +
  labs(x = "number of messages", y = "lapse counts")
```

```
#| label: fig-message_count_1week_byid
#| fig-cap: "Distribution of Message Counts by Subject in the 1-Week Prediction Window"
#| fig-width: 3.5
#| fig-height: 2.5
pred_1week |> 
  group_by(subid, day_start) |> 
  summarize(
    n_messages = sum(!na)
  ) |> 
  group_by(subid) |> 
  summarize(
    mean_messages = mean(n_messages)
  ) |>
  ggplot(aes(x = mean_messages)) +
  geom_histogram(fill = "#97C4C5") +
  labs(x = "averaged number of messages per label", y = "subject counts")
```
